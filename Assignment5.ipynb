```python
class Counter():
    def __init__(self, count, step=1):
        self.count = count
        self.step = step

    def __str__(self):
        s = f"[Count (step: {self.step})] {self.count}"
        return s

    def __repr__(self):
        self.incr()
        s = ""
        return s
    
    def __add__(self, other):
        count = self.count + other
        return Counter(count,self.step) 

    def __sub__(self, other):
        count = self.count - other
        return Counter(count,self.step)
    
    def __gt__(self, other):
        if type(other) == Counter:
            return self.count > other.count
        else:
            return self.count > other
        
    def __lt__(self, other):
        if type(other) == Counter:
            return self.count < other.count
        else:
            return self.count < other
    
    def __eq__(self, other):
        if type(other) == Counter:
            return self.count == other.count
        else:
            return self.count == other
        
    def incr(self):
        self.count += self.step
    
        
c = Counter(10)
d = Counter(10, 2)
```

#### 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
* MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.
    

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

#### 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.
* 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
* 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
> \>>> u = s | t<br>
> \>>> print(u)<br>
> \MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}<br>
* 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
> \>>> u = s & t<br>
> \>>> print(u)<br>
> \MySet: {2, 3}<br>
* 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
> \>>> s = MySet([1, 2, 3])<br>
> \>>> t = MySet([3, 4, 5])<br>
> \>>> u = s - t<br>
> \>>> print(u)<br>
> \MySet: {1, 2}<br>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    def __sub__(self, other):
        ''' '-' 차집합 연산 메소드
        
        MySet 클래스에 self 리스트의 요소를 인자로 넣어 집합을 만든 두개의 집합에서
        self 집합에 other 집합과 겹치는 요소를 제거시킨뒤,
        self 집합을 반환하는 코드이다.
        
        Args: other -> MySet object
        
        return self -> MySet object
        
        '''
        for e in other:
            if e in self:
                self.remove(e)
        return self

    def __and__(self, other):
        ''' '&' 교집합 연산 메소드
        
        MySet 클래스에 차집합 연산을 이용해 self - other 집합의 결과를 뽑아낸뒤
        self 집합에 차집합한 집합의 요소들을 제거하면 최종적으로 교집합이 나오게 된다.
        그뒤 교집합인 self 집합을 반환하는 코드이다.
        
        Args: other -> MySet object
        
        return self -> MySet object
        
        '''
        s = []
        s = self[:]
        for e in other:
            if e in self:
                s.remove(e)
        #print(s)
        for e in s:
            if e in self:
                self.remove(e)
        return self

    def __or__(self, other):
        ''' '|' 합집합 연산 메소드
        
        MySet 클래스에 list 요소를 인자로 넣어 집합을 만든 두개의 집합에서
        self 집합에 other 집합과 겹치는 요소를 제외한 나머지 요소를 
        self 집합에 추가해 반환하는 코드이다.
        
        Args: other -> MySet object
        
        return self -> MySet object
        
        '''
        for e in other:
            if e not in self:
                self.append(e)
        return self
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

    * 파라미터로 받은 단어와 유사도가 동일한 URL이 여러 개이면 해당 URL을 모두 출력하시오.

```python
import requests
import re
import string


class SearchEngine:

    def __init__(self, *args):
        self.urlList = list(args)
        self.count_time = {}
        self.sort_time = []
        self.wordlist = {}

    def __str__(self):
        s = f"url: {self.urlList}"
        return s

    def listUrls(self):
        for url in self.urlList:
            print(url)

    def addUrl(self, url):
        self.urlList.append(url)

    def removeUrl(self, url):
        if url in self.urlList:
            self.urlList.remove(url)

    def getwordFrequency_Each(self, url):
        source = self.getUrlSource(url)
        source = self.removeTags(source)
        source = self.removePunc(source)
        source = self.splitSource(source)
        source = SearchEngine.removeStopEng(source)
        source = SearchEngine.removeStopKor(source)
        return source

    def getWords(self):
        self.count_time = {}

        for url in self.urlList:
            self.countWords(self.getwordFrequency_Each(url))

        SearchEngine.removeStopEng(self.count_time)
        SearchEngine.removeStopKor(self.count_time)

    def getWordsFrequency(self):

        self.getWords()

        self.printWordFrequency()

    def getMaxFreqencyWords(self):

        self.sortWords(True)

        maxFreqWordList = []
        maxFreq = self.sort_time[0][1]
        for word in self.sort_time:
            if word[1] == maxFreq:
                maxFreqWordList.append(word)

        print(maxFreqWordList)

    def getSimilarlist(self, searchWord):
        for url in self.urlList:
            self.wordlist[url] = self.getwordFrequency_Each(url)
            self.wordlist[url] = self.calculateSimilar(self.wordlist[url], searchWord)[0]

    def calculateSimilar(self, wordlist, searchWord):
        findKeyList = searchWord.split()

        SearchEngine.removeStopEng(findKeyList)
        SearchEngine.removeStopKor(findKeyList)

        if len(findKeyList) > 0:
            findkey = findKeyList[0]
        else:
            findkey = ""

        result = []
        default = 100 / len(findkey) if findkey != "" else 1
        for word in wordlist:
            idx = word.find(findkey)
            lendist1 = float(len(findkey) / len(word))
            lendist2 = 1 / (1 + idx) if idx != -1 else 0
            result.append((word, lendist1 * lendist2 * (default * len(findkey)) / 100))

        return sorted(result, key=lambda x: x[1], reverse=True)

    def searchUrlByWord(self, searchWord):
        self.getSimilarlist(searchWord)

        result = sorted(self.wordlist.items(), key=lambda x: x[1][1], reverse=True)

        maxSimilar = []
        maxSim = result[0][1][1]

        for word in result:
            if word[1][1] == maxSim:
                maxSimilar.append(word[0])

        for url in maxSimilar: print(url)

    def countWords(self, source):

        for word in source:
            if word in self.count_time:
                self.count_time[word] += 1
            else:
                self.count_time[word] = 1

    def printWordFrequency(self):
        print(self.count_time)

    def getUrlSource(self, url):
        req = requests.get(url)
        source = req.text
        return source

    def splitSource(self, source):
        source = source.split(' ')
        source = ' '.join(source).split()
        return source

    def sortWords(self, reverse=True):
        self.sort_time = sorted(self.count_time.items(), key=lambda x: x[1], reverse=reverse)

    def removePunc(self, source):
        for rem_str in string.punctuation:
            source = source.replace(rem_str, '')
        return source

    def removeTags(self, source):

        cheak = True

        while cheak:
            numA = source.find('<script')
            numB = source.find('</script>')

            source = source.replace(source[numA:numB + 9], '')

            if numA == -1:
                cheak = False

        cheak = True

        while cheak:
            numA = source.find('<style')
            numB = source.find('</style>')

            source = source.replace(source[numA:numB + 8], '')

            if numA == -1:
                cheak = False

        cheak = True

        while cheak:
            numA = source.find('<url>')
            numB = source.find('</url>')

            source = source.replace(source[numA:numB + 6], '')

            if numA == -1:
                cheak = False

        cheak = True

        while cheak:
            numA = source.find('<!--')
            numB = source.find('-->')

            source = source.replace(source[numA:numB + 3], '')

            if numA == -1:
                cheak = False

        source = re.sub('<[^>]*>', ' ', source)
        source = re.sub('&[^;]*;', ' ', source)
        source = re.sub('@[^@]*@', ' ', source)

        return source

    @staticmethod
    def removeStopEng(wordList):
        reqEng = requests.get(
            "https://raw.githubusercontent.com/leeahwi/k_t_Assignment/main/txt/stop_words_english").text.split('\n')

        if type(wordList) == type({}):
            for word in reqEng:
                if wordList.get(word):
                    del wordList[word]
                if wordList.get(word.title()):
                    del wordList[word.title()]
                if wordList.get(word.upper()):
                    del wordList[word.upper()]
                if wordList.get(word.lower()):
                    del wordList[word.lower()]

        elif type(wordList) == type([]):
            for word in reqEng:
                if word in wordList:
                    wordList.remove(word)
                if word.title() in wordList:
                    wordList.remove(word.title())
                if word.upper() in wordList:
                    wordList.remove(word.upper())
                if word.lower() in wordList:
                    wordList.remove(word.lower())

        return wordList

    @staticmethod
    def removeStopKor(wordList):
        reqKor = requests.get(
            "https://raw.githubusercontent.com/leeahwi/k_t_Assignment/main/txt/stop_words_korean.txt").text.split('\n')

        if type(wordList) == type({}):
            for word in reqKor:
                if wordList.get(word.strip()):
                    del wordList[word.strip()]

        elif type(wordList) == type([]):
            for word in reqKor:
                if word.strip() in wordList:
                    wordList.remove(word.strip())

        return wordList
```

#### 2. 위 1번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

```python
class SearchEngineWithOrderedWebWords(SearchEngine):

    def __iter__(self):
        self.count = 0
        return self

    def __next__(self):
        if self.count < len(self.sort_time):
            next = self.sort_time[self.count]
            self.count += 1
            return next
        else:
            raise StopIteration

    def getWordsFrequency(self,reverse=False):
        super(SearchEngineWithOrderedWebWords,self).getWords()
        self.sortWords(not reverse)
        print(self.sort_time)
```

* 리팩토링이 생각보다도 시간이 걸렸었다. 각 메소드들이 하나의 기능을 가지게 만드는것을 생각하며 코드를 짜다보니, 생각보다도 각 메소드들을 모아 원하는 기능을 만드는 것에 시간이 걸렸다. 각 메소드들이 서로 유기적으로 연결되어, 합쳐져서 작동을 해야했으며 중간중간 클래스의 상태를 원치 않은 방향으로 변경시켜 원하는 결과가 안나오기도해 애를 많이 먹었던것 같다.
* 그래도 시행착오를 거치며 리팩토링을 거친 결과는 이전과는 비교안될만큼 간결하고 가독성이 높은 코드가 결과로 나왔으며, 추가적으로 매직 메소드들을 활용해 더욱 해당 프로젝트의 활용도를 높일 수 있었다고 생각한다.
