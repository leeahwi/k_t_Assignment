* 1) addall(list) 함수 <br>(리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)<br>

* [참고 2]: sum() 등의 내장함수 사용 가능함<br><br>

```python
def addall(list):
    sum = 0
    for num in list:
        sum += num
    
    print(sum)

addall([1])
addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
def addall(list):
    print(sum([num for num in list]))
    
addall([1])
addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
def addallodd(list):
    sum = 0
    for num in list:
        if num%2==1:
            sum += num
    print(sum)
    
addallodd([1])
addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
def addallodd(list):
    print(sum([num for num in list if num%2 == 1]))
    
addallodd([1])
addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}

#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
# 2)
sorted_key = sorted(d.items())
sorted_keyreverse = sorted_key[::-1]
for num in range(len(sorted_keyreverse)):
    print(sorted_keyreverse[num])
```

##### [문제 7. 설명]
* 2) sorted 함수를 이용해 key의 알파벳 오름차 순서로 정렬된 key와 value를 튜플 형태로 요소로 넣은 sorted_key 리스트를 얻어내고 슬라이싱을 통해 역순으로 들어간 sorted_keyreverse를 얻어내었다. 그다음 반복문을 통해 sorted_keyreverse 리스트를 순서대로 출력하였다. 앞에서 말했듯이 출력은 튜플 형태로 출력된다.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}

#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
# 3)
sorted_value = sorted(d.items(), key=lambda x : x[1])
for num in range(len(sorted_value)):
    print(sorted_value[num])
```

* length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수
* Strange라는 숫자를 아래과 같이 정의한다.
    * 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
    * length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
        * X는 length(X)로 정확히 나누어 떨어진다.
        * X / length(X) 가 strnage이다.  
          
          
* 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
* isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.

```python
def isStrange(num):
    
    def length(X):
        length_len = 1
        while X/10 >= 1:
            X = X/10
            length_len += 1
        return length_len
    
    
    length_len = length(num)
    
    if num >= 0 and length_len == 1:
        return True
    
    elif length_len > 1:
        
        if num%length_len == 0 and isStrange(num/length_len):
            return True
    
    return False
    
print(isStrange(10))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
```

#### 9. 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.
* 1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고, ith 그룹에는 임의의 팀원 수 ai 명이 있다.
하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.
하지만, 같은 그룹에 속한 팀원들은 분리되기를 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.
또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.
마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.    


* 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수 solve(a) 를 작성하시오.
* [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

```python
def select(a, n):
    sum = 0
    for num in a:
        if(sum + num > n):
            return False
        elif(sum + num == n):
            sum = 0
        else:
            sum += num
            
    return sum==0
    

def solve(a):
    b = sorted(a)

    max = b[-1]

    n = sum([num for num in a])

    div_List = []

    for i in range(1, int(n ** (1 / 2)) + 1):
        if (n % i == 0):
            div_List.append(i)
            if (i != (n // i)):
                div_List.append(n // i)

    div_List.sort()
    divisors_sorted = [num for num in div_List if num>= max]
    
    divisors_result = [num_i for num_i in divisors_sorted if select(a, num_i)]

    return divisors_result

print(solve([9999] * 10**5))
print(solve([1, 2, 1, 1, 1, 2, 1, 3]))
print(solve([1, 1, 1, 1]))
print(solve([2, 3, 3, 2, 4, 3]))
```

### [Incremental Project 문제]
#### 이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오.
* 요구사항 1. CSS와 Javascript 내용도 모두 제거하시오 (이전 숙제에서 이미 제외하였다면 본 요구사항 무시)
* 요구사항 2. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.
* 요구사항 3. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.
* 요구사항 4. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.
* 요구사항 5. 다음 5개의 사이트에 대한 단어 출현빈도 결과를 출력하시오.

```python
import requests
import re
import string

def print_pagewords(url):
    req = requests.get(url)

    source = req.text

    cheak = True

    while cheak:
        numA = source.find('<script')
        numB = source.find('</script>')

        source = source.replace(source[numA:numB+9], '')

        if numA == -1:
            cheak = False

    cheak = True

    while cheak:
        numA = source.find('<style')
        numB = source.find('</style>')

        source = source.replace(source[numA:numB + 8], '')

        if numA == -1:
            cheak = False

    cheak = True

    while cheak:
        numA = source.find('<url>')
        numB = source.find('</url>')

        source = source.replace(source[numA:numB + 6], '')

        if numA == -1:
            cheak = False
            
    cheak = True
            
    while cheak:
        numA = source.find('<!--')
        numB = source.find('-->')

        source = source.replace(source[numA:numB + 3], '')

        if numA == -1:
            cheak = False

    cheak = True

    source = re.sub('<[^>]*>', ' ', source)
    source = re.sub('&[^;]*;', ' ', source)
    source = re.sub('@[^@]*@', ' ', source)
    
    
    for rem_str in string.punctuation:
            source = source.replace(rem_str, '')
    
    source = source.split(' ')        
    source = ' '.join(source).split()
    
    count_time = {}

    for word in source:
        if word in count_time:
            count_time[word] += 1
        else:
            count_time[word] = 1
        
    print(count_time)


print_pagewords('https://www.daum.net/')
print()
print_pagewords('https://www.koreatech.ac.kr')
print()
print_pagewords('https://www.naver.com/')
print()
print_pagewords('http://cse.koreatech.ac.kr')
print()
print_pagewords('http://www.nytimes.com')
```
